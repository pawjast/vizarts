# Recreation of the album cover: Black Keys - Turn Blue

[original cover](https://upload.wikimedia.org/wikipedia/en/b/bb/Black_Keys_Turn_Blue_album_cover.png)

## Imports

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
```

## Data & Plot

```python
# Parameters
nr_of_points = 1000
phi = np.linspace(0, 2.1*np.pi, nr_of_points)
r = np.linspace(.0, 1, nr_of_points)
r_multiplier = 1.065
phi_multiplier = 0.1

fig, ax = plt.subplots(
    figsize=(10, 10),
    facecolor="#0174afff"
)
fig.suptitle(
    "Black Keys - Turn Blue",
    fontsize=35,
    fontfamily="serif",
    weight="bold",
    color="white"
)
ax.set_aspect("equal")
ax.set_facecolor("#e10a79ff")

# ax.plot(x, y)
# ax.plot(xx, yy)
for delta_phi in np.arange(0, 2*np.pi, np.pi/12):
    x = r*np.cos(phi+delta_phi) * np.exp(phi*0.105)
    y = -r*np.sin(phi+delta_phi) * np.exp(phi*0.105)

    xx = r_multiplier*r*np.cos(phi+delta_phi) * np.exp(phi*0.1)
    yy = -r_multiplier*r*np.sin(phi+delta_phi) * np.exp(phi*0.1)

    ax.fill(
        np.concatenate((x, xx[::-1])),
        np.concatenate((y, yy[::-1])),
        color="#007dbcff"
    )

# Set axis limits
lim = 1.34
ax.set_xlim(-lim, lim)
ax.set_ylim(-lim, lim)

# Add circle
patch = patches.Circle(
    xy=(0,0),
    radius=.368,
    facecolor="#737373ff"
)
ax.add_patch(patch)

# Add text
text = ax.text(
    x=0,
    y=0,
    s="THE\nBLACK KEYS\nTURN BLUE",
    color="#636363ff",
    ha="center",
    va="center",
    fontsize=23,
    fontweight="black"
)

# Remove axis ticks and tick labels
ax.tick_params(
    axis="both",
    which="major",
    left=False,
    labelleft=False,
    bottom=False,
    labelbottom=False
)

# Remove spines
for position in ["top", "right", "bottom", "left"]:
    ax.spines[position].set_visible(False)

# Footer
ax.text(
    1, -0.06,
    "by @pawjast\nmade with @matplotlib \u2665",
    transform = ax.transAxes,
    horizontalalignment="right",
    verticalalignment="bottom",
    color="white",
    fontsize=16,
    family="monospace",
    weight="medium"
)

plt.tight_layout()

# plt.savefig(
#     fname="vizart_album_cover_black_keys.png",
#     format="png",
#     dpi=125,
# );
```
