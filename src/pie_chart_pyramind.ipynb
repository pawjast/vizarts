## Data

```python
data = {
    "Sky": 12,
    "Sunny side of pyramid": 3,
    "Shady side of pyramid": 1,
 }
```

## Plot

### Settings

```python
plt.rcParams.update(
    {
        "font.family": "Roboto",
        "font.size": 16,
        "hatch.linewidth": 3,
        "axes.spines.top": False,
        "axes.spines.right": False,
        # "axes.spines.left": False,
        # "axes.spines.bottom": False,
    }
)

col_blue = "#3e6390"
col_yellow = "#d0a15d"
col_brown = "#5d3c29"
```

### Original colours

```python

fig, ax = plt.subplots(
    figsize=(9, 6),
    facecolor=("whitesmoke", 0.5),
    layout="constrained",
    dpi=150
)
fig.suptitle(
    "Pyramid chart",
    fontsize=40,
    weight="bold"
)

ax.pie(
    x=data.values(),
    colors=[col_blue, col_yellow, col_brown],
    startangle=-40,
    radius=1.1,
)

# Create mock "handles" (circles) matching your pie colors
blue_circle = mlines.Line2D(
    [], [],
    color=col_blue,
    marker="o",
    markersize=45,
    linestyle="None",
)
yellow_circle = mlines.Line2D(
    [], [],
    color=col_yellow,
    marker="o",
    markersize=45,
    linestyle="None",
)
brown_circle = mlines.Line2D(
    [], [],
    color=col_brown,
    marker="o",
    markersize=45,
    linestyle="None",
)
ax.legend(
    handles=[blue_circle, yellow_circle, brown_circle],
    labels=data.keys(),
    labelspacing=2.5,
    bbox_to_anchor=(1, 0.5),
    loc="center left",
    frameon=False,
)


# Footer
ax.text(
    1,
    0.025,
    "by @pawjast\nmade with @matplotlib \u2665",
    transform=fig.transFigure,
    horizontalalignment="right",
    verticalalignment="bottom",
    color="#9a9996",
    fontsize=16,
    family="monospace",
    weight="medium",
)

# plt.savefig(
#     fname="name.png",
#     format="png",
#     dpi=125,
#     metadata=metadata_dict
# );
```

### My colours

```python
col_blue = "#6aadff"
col_yellow = "#ffda82"
col_brown = "#ec8245"
```

```python
fig, ax = plt.subplots(
    figsize=(9, 6),
    facecolor=col_blue,
    layout="constrained",
    dpi=150
)
fig.suptitle(
    "Pyramid chart",
    fontsize=40,
    weight="bold",
    color="white",
)

ax.pie(
    x=data.values(),
    colors=[col_blue, col_yellow, col_brown],
    startangle=-40,
    radius=1.1,
    wedgeprops=dict(
        linewidth=2,
        edgecolor="white"
    ),
)

# Create mock "handles" (circles) matching your pie colors
blue_circle = mlines.Line2D(
    [],
    [],
    color=col_blue,
    marker="o",
    markersize=45,
    markeredgecolor="white",
    linestyle="None",
)
yellow_circle = mlines.Line2D(
    [],
    [],
    color=col_yellow,
    marker="o",
    markersize=45,
    markeredgecolor="white",
    linestyle="None",
)
brown_circle = mlines.Line2D(
    [],
    [],
    color=col_brown,
    marker="o",
    markersize=45,
    markeredgecolor="white",
    linestyle="None",
)
ax.legend(
    handles=[blue_circle, yellow_circle, brown_circle],
    labels=data.keys(),
    labelspacing=2.5,
    labelcolor="white",
    bbox_to_anchor=(1, 0.5),
    loc="center left",
    frameon=False,
)


# Footer
ax.text(
    1,
    0.025,
    "by @pawjast\nmade with @matplotlib \u2665",
    transform=fig.transFigure,
    horizontalalignment="right",
    verticalalignment="bottom",
    color="white",
    fontsize=16,
    family="monospace",
    weight="medium",
)

# plt.savefig(
#     fname="name.png",
#     format="png",
#     dpi=125,
#     metadata=metadata_dict
# );
```
