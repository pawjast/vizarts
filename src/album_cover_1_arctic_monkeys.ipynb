```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# DEV
from datetime import date

metadata_dict = {
    "Title": "Vizarts for Twitter/X",
    "Author": "Pawel Jastrzebski (pawjast)",
    "Description": "https://medium.com/@pawjast https://twitter.com/pawjast https://github.com/pawjast www.linkedin.com/in/pawjast https://www.youtube.com/@pawjast https://pawjast.substack.com/",
    "Creation Time": f"{date.today():%d %B %Y}",
    "Source": "https://twitter.com/pawjast"
}
```

```python
nr_of_samples = 1000

t = np.linspace(0, 23, nr_of_samples)


A = np.sin(np.linspace(0, 2*np.pi, nr_of_samples))
y = A*np.sin(2*np.pi*t)
```

```python
fig, ax = plt.subplots(
    figsize=(10, 10),
    facecolor="white"
)

fig.suptitle(
    "Arctic Monkeys - AM",
    fontsize=35,
    fontfamily="serif",
    weight="bold"
)

ax.plot(
    t, y,
    color="white",
    linewidth=4
)

ax.set_facecolor("black")

ax.set_xlim(0.8, 22.15)
ax.set_ylim(-3.25, 3.25)

# Remove `y` axis ticks and tick labels
ax.tick_params(
    axis="both",
    which="major",
    left=False,
    labelleft=False,
    bottom=False,
    labelbottom=False
)

# Remove spines
for position in ["top", "right", "bottom", "left"]:
    ax.spines[position].set_visible(False)

# Footer
ax.text(
    1, -0.04,
    "by pawjast @ Twitter/X",
    transform = ax.transAxes,
    horizontalalignment="right",
    verticalalignment="bottom",
    color="#9a9996",
    fontsize=16,
    family="monospace",
    weight="medium"
)

plt.tight_layout()

plt.savefig(
    fname="1.png",
    format="png",
    dpi=125,
    metadata=metadata_dict
);
```
