# Recreation of the album cover: Ed Sheeran - X

[original cover](https://en.wikipedia.org/wiki/X_(Ed_Sheeran_album)#/media/File:X_cover.png)

## Imports

```python
import matplotlib.pyplot as plt
from matplotlib import font_manager
```

```python
# And system font so matplotlib sees them
for font in font_manager.findSystemFonts(fontpaths=None, fontext='ttf'):
    font_manager.fontManager.addfont(font)
```

## Data & Plot

```python
fig, ax = plt.subplots(
    figsize=(10, 10),
    facecolor="white",
    layout="constrained"
)
fig.suptitle(
    "Ed Sheeran - X",
    fontsize=35,
    fontfamily="serif",
    weight="bold",
    color="black"
)
ax.set_aspect("equal")
ax.set_facecolor("#1ccb55ff")

with plt.rc_context({'font.family': "Poppins"}):
    ax.text(
        x=0,
        y=0.025,
        s="X",
        weight="heavy",
        ha="center",
        va="center_baseline",
        fontsize=600
    )

ax.set_xlim(-1, 1)
ax.set_ylim(-1, 1)

# Remove axis ticks and tick labels
ax.tick_params(
    axis="both",
    which="major",
    left=False,
    labelleft=False,
    bottom=False,
    labelbottom=False
)

# Remove spines
for position in ["top", "right", "bottom", "left"]:
    ax.spines[position].set_visible(False)

# Footer
ax.text(
    1, -0.1,
    "by @pawjast\nmade with @matplotlib \u2665",
    transform = ax.transAxes,
    horizontalalignment="right",
    verticalalignment="bottom",
    color="#9a9996",
    fontsize=16,
    family="monospace",
    weight="medium"
);

# plt.savefig(
#     fname="vizart_edsheeran_x.png",
#     format="png",
#     dpi=125,
# );
```
