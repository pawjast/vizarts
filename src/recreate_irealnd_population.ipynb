## Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager
```

## Data

```python
df = pd.DataFrame.from_dict(
    {
        1821: {"population": 6.8, "note": "First official census"},
        1831: {"population": 7.8, "note": "Rapid pre-famine growth"},
        1841: {"population": 8.2, "note": "Historical peak before the Great Famine"},
        1851: {"population": 6.5, "note": "Impact of Great Famine and mass emigration"},
        1861: {"population": 5.8, "note": "Continued post-famine decline"},
        1871: {"population": 5.4, "note": "Ongoing high emigration rates"},
        1881: {"population": 5.2, "note": "Land War period"},
        1891: {"population": 4.7, "note": "Population continues to contract"},
        1901: {"population": 4.5, "note": "Turn of the century"},
        1911: {"population": 4.4, "note": "Last census before Irish independence"},
        1926: {"population": 4.2, "note": "First census of the Irish Free State"},
        1936: {"population": 4.2, "note": "Period of economic stagnation"},
        1946: {"population": 4.3, "note": "Post-WWII / The Emergency"},
        1951: {"population": 4.3, "note": "Brief period of stability"},
        1961: {"population": 4.2, "note": "All-time population low point"},
        1971: {"population": 4.5, "note": "Beginning of modern recovery"},
        1981: {"population": 5.0, "note": "Significant growth despite economic strife"},
        1991: {"population": 5.2, "note": "Pre-Celtic Tiger era"},
        2002: {"population": 5.6, "note": "Rapid expansion during Celtic Tiger"},
        2011: {"population": 6.4, "note": "Post-2008 crash recovery"},
        2021: {"population": 7.0, "note": "Strongest growth in over a century"},
        2025: {"population": 7.4, "note": "Current estimate (Combined ROI & NI)"},
    },
    orient="index"
)

df
```

## Add fonts

```python
# Add system fonts for Matplotlib to see them
for font in font_manager.findSystemFonts(fontpaths=None, fontext="ttf"):
    font_manager.fontManager.addfont(font)
```

```python
font_manager.get_font_names()
```

## Plot settings

```python
my_custom_grey = "#5e5c64"
```

```python
plt.rcParams.update(
    {
        "font.family": "DejaVu Serif",
        "font.size": 18,
        "axes.spines.right": False,
        "axes.spines.top": False,
        "axes.titlelocation": "left",
        "axes.titlepad": 20,
        "text.color": my_custom_grey,
        "axes.labelcolor": my_custom_grey,
        "xtick.color": my_custom_grey,
        "ytick.color": my_custom_grey,
    }
)
```

## Plot

```python
fig, ax  = plt.subplots(
    figsize=(16, 10),
    facecolor="whitesmoke"
)

ax.set_facecolor("whitesmoke")

ax.set_title(
    "Population of the island of Ireland",
    fontsize=30,
    weight="bold",
    loc="left",
)

ax.plot(
    df.population,
    marker="o",
    color="black"
)
ax.fill_between(
    x=df.index,
    y1=df.population,
    color="gainsboro"
)
ax.axhline(
    y=df.population.max(),
    color=my_custom_grey,
    linestyle=":"
)
ax.axhline(
    y=df.population.min(),
    color=my_custom_grey,
    linestyle=":"
)

ax.text(
    x=df[df.population == df.population.max()].population.index.values[0],
    y=df[df.population == df.population.max()].population.values[0] + 0.2,
    s=f"{df[df.population == df.population.max()].population.values[0]}M ({df[df.population == df.population.max()].population.index.values[0]})",
    va="bottom",
    ha="center",
)
ax.text(
    x=df[df.population == df.population.min()].population.index.values[0],
    y=df[df.population == df.population.min()].population.values[0] - 0.2,
    s=f"{df[df.population == df.population.min()].population.values[0]}M ({df[df.population == df.population.min()].population.index.values[0]})",
    va="top",
    ha="center",
)

ax.text(
    x=df.iloc[-1, :].name,
    y=df.iloc[-1, :].population + 0.2,
    s=f"{df.iloc[-1, :].population}M ({df.iloc[-1, :].name})",
    va="bottom",
    ha="center",
)

ax.set_ylabel("Population")
ax.set_ylim(0, 10)
ax.set_yticks(
    ticks=range(1, 9),
    labels=[f"{y}M" for y in range(1, 9)]
)

ax.text(
    1,
    -0.13,
    "by @pawjast\nmade with @matplotlib \u2665",
    transform=ax.transAxes,
    horizontalalignment="right",
    verticalalignment="bottom",
    color="#9a9996",
    fontsize=16,
    family="monospace",
    weight="medium",
);
```
