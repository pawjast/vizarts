## Helper functions

### Plot data

```python
def plot_presidents(group: str, df: pd.DataFrame, ax: Axes) -> None:
    """Plot the data"""
    # set colour
    if group == "Democratic":
        col_val="blue"
    elif group == "Republican":
        col_val="red"
    else:
        col_val="green"

    # group by status: needed for setting markers
    grouped_by_status = df.groupby(by="status")

    for name, values in grouped_by_status:
        if name == "assassinated":
            marker_val = "x"
        else:
            marker_val = "."


        ax.plot(
            values["date_of_inauguration"],
            values["age_at_inauguration"],
            markerfacecolor=col_val,
            markeredgecolor=col_val,
            marker=marker_val,
            markersize=8,
            linestyle="",
            label=name
        )
```

### set X and Y ticks

```python
def set_ticks(df: pd.DataFrame, ax: Axes) -> None:
    mean = df["age_at_inauguration"].mean()
    std = round(df["age_at_inauguration"].std(), 1)

    ax.set_yticks(
        [
            mean - 3*std,
            mean - 2*std,
            mean - std,
            # mean, 
            mean + std,
            mean + 2*std,
            mean + 3*std
        ]
    )

    x_vals = pd.date_range(
        start="1/1/1800",
        periods=5,
        freq="50YS"
    )
    ax.set_xticks(
        ticks=x_vals,
        labels=[f"{date: %Y}" for date in x_vals]
    )
```

### Mark alive presidents

```python
def mark_alive_presidents(start, ax: Axes, end=None) -> None:

    if end:
        # Two dates
        x1 = start
        x2 = end
    else:
        # One date
        x1 = start
        x2 = start

    ax.fill_betweenx(
        y=[0, 80],
        x1=x1 - pd.Timedelta(days=1.5*365),
        x2=x2 + pd.Timedelta(days=1.5*365),
        facecolor="#deddda",
        alpha=0.35
    )
```

### Make legend

```python
def make_legend(ax: Axes) -> None:
    elements = [
        Line2D(
            [0], [0],
            linestyle="",
            marker=".",
            markersize=8,
            markerfacecolor="blue",
            markeredgecolor="blue",
            label="Democrats"
        ),
        Line2D(
            [0], [0],
            linestyle="",
            marker=".",
            markersize=8,
            markerfacecolor="red",
            markeredgecolor="red",
            label="Republicans"
        ),
        Line2D(
            [0], [0],
            linestyle="",
            marker=".",
            markersize=8,
            markerfacecolor="green",
            markeredgecolor="green",
            label="Other"
        ),
        Line2D(
            [0], [0],
            linestyle="",
            marker="x",
            markersize=8,
            markerfacecolor="#9a9996",
            markeredgecolor="#9a9996",
            label="assassinated"
        ),
        Patch(
            facecolor="#deddda",
            alpha=0.35,
            label="still alive"
        )
    ]

    ax.legend(
        handles=elements,
        frameon=False,
        ncols=5,
        bbox_to_anchor=(0., .98, 1., .1),
        loc="lower left"
    )
```

### Make annotation

```python
def make_annotation(df, ax: Axes, xy=(0.5, 0.5), add_text=None) -> None:
    ax.annotate(
        text=f"{add_text} {df['full_name'].values[0]}",
        xy=(
            df["date_of_inauguration"].values[0],
            df["age_at_inauguration"].values[0],
        ),
        xycoords="data",
        xytext=xy,
        textcoords=ax.transAxes,
        arrowprops=dict(
            arrowstyle="-",
            connectionstyle="arc3,rad=-0.2",
            shrinkA=15, 
            shrinkB=10,
            color="#5e5c64"
        )
    )
```

## Plot

```python
fig, ax = plt.subplots(
    figsize=(16, 10)
)

ax.set_title(
    "Presidents of the United States\nby age at inauguration date",
    fontsize=25,
    weight="bold"
)

for party, values in grouped_by_party:
    plot_presidents(
        group=party,
        df=values,
        ax=ax
    )

# Mean age
ax.axhline(
    df["age_at_inauguration"].mean(),
    lw=1.25,
    color="purple"
)

# Alive?
mark_alive_presidents(
    start=df["date_of_inauguration"].iloc[38],
    ax=ax
)
mark_alive_presidents(
    start=df["date_of_inauguration"].iloc[41],
    end=df["date_of_inauguration"].iloc[45],
    ax=ax
)

set_ticks(
    df=df,
    ax=ax
)

ax.set_xlim(
    pd.Timestamp("1775-1-1"),
    pd.Timestamp("2025-1-1")
)
ax.set_ylim(0, 80)

ax.text(
    ax.get_xlim()[0],
    df["age_at_inauguration"].mean(),
    f"{df['age_at_inauguration'].mean(): .2f}",
    va="center",
    ha="right",
    color="purple"
)

make_annotation(
    df=df[df["age_at_inauguration"] == max(df["age_at_inauguration"])],
    ax=ax,
    xy=(0.55, 0.885),
    add_text="The oldest: "
)
make_annotation(
    df=df[df["age_at_inauguration"] == min(df["age_at_inauguration"])],
    ax=ax,
    xy=(0.24, 0.43),
    add_text="The youngest: "
)

ax.grid(
    axis="both",
    color="#deddda"
)

make_legend(ax=ax)

# Footer
ax.text(
    0, -0.13,
    "Data: generated with chatGPT\n",
    transform = ax.transAxes,
    horizontalalignment="left",
    verticalalignment="bottom",
    color="#9a9996",
    fontsize=16,
    family="monospace",
    weight="medium"
)
ax.text(
    1, -0.13,
    "by @pawjast\nmade with @matplotlib \u2665",
    transform = ax.transAxes,
    horizontalalignment="right",
    verticalalignment="bottom",
    color="#9a9996",
    fontsize=16,
    family="monospace",
    weight="medium"
);

plt.savefig(
    fname="us_presidents_age_at_inauguration.png",
    format="png",
    dpi=125,
    metadata=metadata_dict
);
```
