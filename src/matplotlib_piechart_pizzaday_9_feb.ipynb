## Data

```python
data = {
    "Dec": 1,
    "Nov": 1,
    "Oct": 1,
    "Sep": 1,
    "Aug": 1,
    "Jul": 1,
    "Jun": 1,
    "May": 1,
    "Apr": 1,
    "Mar": 1,
    "Feb": 1,
    "Jan": 1
 }
```

```python
labels = [f"9th of {key}" if key == "Feb" else "" for key in data.keys()]
labels
```

## Plot

```python
plt.rcParams.update(
    {
        "font.family": "Roboto",
        "font.size": 22,
        "hatch.linewidth": 3,
        "axes.spines.top": False,
        "axes.spines.right": False,
        # "axes.spines.left": False,
        # "axes.spines.bottom": False,
    }
)
```

### Settings

```python
explode_val = np.zeros(len(data))
explode_val[10] = 0.2
```

```python

fig, ax = plt.subplots(
    figsize=(9, 6),
    facecolor=("white", 0.5),
    layout="constrained",
    dpi=150
)
fig.suptitle(
    "International Pizza Day",
    fontsize=40,
    weight="bold"
)


# Mozzarella and tomato base
ax.pie(
    x=data.values(),
    colors=["#ec4511ff"],
    explode=explode_val,
    startangle=90,
    wedgeprops=dict(
        linewidth=1,
        edgecolor="#f2e0c6ff"
    ),
    textprops=dict(
        size=30
    ),
    hatch=["O."]
)

# Slice edges
ax.pie(
    x=data.values(),
    colors=[(0, 0, 0, 0)],
    explode=explode_val,
    startangle=90,
    wedgeprops=dict(
        linewidth=1,
        edgecolor="brown"
    ),
    textprops=dict(
        size=30
    ),
)

# Crust
ax.pie(
    x=data.values(),
    colors=["#eda948ff"],
    labels=labels,
    explode=explode_val,
    startangle=90,
    wedgeprops=dict(
        width=0.12,
        linewidth=1,
        edgecolor=("brown")
    )
)


# Footer
ax.text(
    1, 0,
    "by @pawjast\nmade with @matplotlib \u2665",
    transform = ax.transAxes,
    horizontalalignment="right",
    verticalalignment="bottom",
    color="#9a9996",
    fontsize=16,
    family="monospace",
    weight="medium"
)

# plt.savefig(
#     fname="name.png",
#     format="png",
#     dpi=125,
#     metadata=metadata_dict
# );
```
