## Helper functions

### Plot data

```python
def plot_presidents(group: str, df: pd.DataFrame, ax: Axes) -> None:
    """Plot the data"""
    # set colour
    if group == "Democratic":
        col_val="blue"
    elif group == "Republican":
        col_val="red"
    else:
        col_val="green"

    # group by status: needed for setting markers
    grouped_by_status = df.groupby(by="status")

    for name, values in grouped_by_status:
        if name == "assassinated":
            marker_val = "x"
        else:
            marker_val = "."


        ax.plot(
            values["date_of_inauguration"],
            values["age_at_inauguration"],
            markerfacecolor=col_val,
            markeredgecolor=col_val,
            marker=marker_val,
            markersize=8,
            linestyle="",
            label=name
        )
```

### set X and Y ticks

```python
def set_ticks(df: pd.DataFrame, ax: Axes) -> None:
    mean = round(df["age_at_inauguration"].mean())
    std = round(df["age_at_inauguration"].std())

    ax.set_yticks(
        [
                mean - 3*std,
                mean - 2*std,
                mean - std,
                # mean, 
                mean + std,
                mean + 2*std,
                mean + 3*std
        ]
    )

    x_vals = pd.date_range(
        start="1/1/1800",
        periods=5,
        freq="50YS"
    )
    ax.set_xticks(
        ticks=x_vals,
        labels=[f"{date: %Y}" for date in x_vals]
    )
```

### Mark alive presidents

```python
def mark_alive_presidents(start, ax: Axes, end=None) -> None:

    if end:
        # Two dates
        x1 = start
        x2 = end
    else:
        # One date
        x1 = start
        x2 = start

    ax.fill_betweenx(
        y=[0, 80],
        x1=x1 - pd.Timedelta(days=1.5*365),
        x2=x2 + pd.Timedelta(days=1.5*365),
        facecolor="#deddda",
        alpha=0.35
    )
```

### Make legend

```python
def make_legend(ax: Axes):
    """This function manually add elements of the legend"""
    elements = [
        # Line2D(
        #     [0], [0],
        #     linestyle="",
        #     marker=".",
        #     markersize=8,
        #     markerfacecolor="blue",
        #     markeredgecolor="blue",
        #     label="Democrats"
        # ),
        # Line2D(
        #     [0], [0],
        #     linestyle="",
        #     marker=".",
        #     markersize=8,
        #     markerfacecolor="red",
        #     markeredgecolor="red",
        #     label="Republicans"
        # ),
        # Line2D(
        #     [0], [0],
        #     linestyle="",
        #     marker=".",
        #     markersize=8,
        #     markerfacecolor="green",
        #     markeredgecolor="green",
        #     label="Other"
        # ),
        Line2D(
            [0], [0],
            linestyle="",
            marker="x",
            markersize=8,
            markerfacecolor="#9a9996",
            markeredgecolor="#9a9996",
            label="assassinated"
        ),
        Patch(
            facecolor="#deddda",
            alpha=0.35,
            label="still alive"
        )
    ]

    ax.legend(
        handles=elements,
        frameon=False,
        ncols=2,
        bbox_to_anchor=(0., .98, 1., .1),
        loc="lower center"
    )
```

### Make annotation

```python
def make_annotation(df, ax: Axes, xy=(0.5, 0.5), add_text=None) -> None:
    ax.annotate(
        text=f"{add_text} {df['full_name'].values[0]}",
        xy=(
            df["date_of_inauguration"].values[0],
            df["age_at_inauguration"].values[0],
        ),
        xycoords="data",
        xytext=xy,
        textcoords=ax.transAxes,
        arrowprops=dict(
            arrowstyle="-",
            connectionstyle="arc3,rad=-0.2",
            shrinkA=15, 
            shrinkB=10,
            color="#5e5c64"
        )
    )
```
