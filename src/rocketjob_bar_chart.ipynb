## Imports

```python
import matplotlib.pyplot as plt
from matplotlib import font_manager
import pandas as pd
```

### The custom fonts

Get more from [Google Fonts](https://fonts.google.com/)

```python
# Check the fonts Matplotlib can find
font_manager.get_font_names()
```

```python
# List all OS fonts
font_manager.findSystemFonts(fontpaths=None, fontext='ttf')
```

```python
# Scan for OS fonts and add them to the font manager so Matplotlib recognizes them
for font in font_manager.findSystemFonts(fontpaths=None, fontext='ttf'):
    font_manager.fontManager.addfont(font)
```

## Data

```python
df = pd.DataFrame.from_dict(
    {
        "Styczeń 2024": [18000, 60000],
        "Styczeń 2025": [24000, 56000],
        "Styczeń 2026": [30500, 52000],
    },
    orient="index",
    columns=[
        "rocketjobs.pl + justjoin.it",
        "Grupa Pracuj",
    ]
)

df
```

## Matplotlib global settings

```python
text_colour = "#5e5c64"
bar_colours = [
    "cyan",
    "navy"
]

plt.rcParams.update(
    {
        "font.family": "Lato",
        "font.size": 16,
        "text.color": text_colour,
        "xtick.labelcolor": text_colour,
        "xtick.bottom": False  # Set to True if you want to see the ticks
    }
)
```

## Plot

```python
# Create figure and axis
fig, ax = plt.subplots(
    figsize=(12,6),
    facecolor="white"
)

# Figure title
fig.suptitle(
    "Liczba Ogłoszeń",
    fontsize=25,
    weight="bold",
)

# Plot the data
df.plot(
    kind="bar",
    ax=ax,
    color=bar_colours
)

for container, col in zip(ax.containers, bar_colours):
    ax.bar_label(container, padding=3, color=col)

# Set and rotate the X axis tick labels
ax.set_xticks(
    ticks=range(len(df)),
    labels=df.index,
    rotation=0,
    ha="center",
)

# Move legend
ax.legend(
    loc="lower center",
    bbox_to_anchor=(0.5, 1.02),
    ncol=2,
    frameon=False,
)

# Remove `y` axis ticks and tick labels
ax.tick_params(
    axis="y",
    which="major",
    left=False,
    labelleft=False
)

# Remove spines
for position in ["left", "top", "right"]:
    ax.spines[position].set_visible(False)

plt.tight_layout()

# plt.savefig(
#     fname="plot.png",
#     format="png",
#     dpi=125
# );
```
