## Get data

```python
gdf = gpd.read_file(
    gpd.datasets.get_path("naturalearth_lowres")
)

gdf
```

```python
gdf.crs
```

```python
df = pd.read_html(
    io="https://www.worldometers.info/geography/how-many-countries-in-asia/"
)

df = df[0]
df
```

### Asia data

```python
df_asia = pd.DataFrame(
    np.zeros(
        shape=(
            len(df),
            len(string.ascii_uppercase)
        ),
        dtype=np.int8
    ),
    index=df["Country"].str.upper(),
    columns=list(string.ascii_uppercase)
)

df_asia
```

```python
df_asia.info()
```

## Calculate letter distribution

```python
for r in df_asia.iterrows():
    df_asia.loc[r[0], :] = [1 if letter in r[1].name else 0 for letter in string.ascii_uppercase]
```

```python
df_asia
```

```python
df_asia.info()
```

## Plot

```python
mosaic = [
    ["map"]*6,
    ["map"]*6,
    ["letters", "letters", "letters", "hist", "bars", "bars"]
]

bright_green = "#8ff0a4"
dark_green = "#26a269"

my_cmap = ListedColormap(
    ["white", bright_green]
)

x_bar_chart_values = sorted(list(set(df_asia.sum().values)))


rcParams.update(
    {
        "font.size": 14,
    }
)
```

```python
fig, axs = plt.subplot_mosaic(
    mosaic=mosaic,
    figsize=(20, 18),
    layout="constrained"
)

# Map plot
axs["map"].set_title(
    "Map of the World with Asia highlighted",
    fontsize=20,
    weight="bold"
)
gdf.to_crs('ESRI:54012').boundary.plot(
    ax=axs["map"],
    edgecolor=dark_green,
    lw=1
)
gdf[gdf["continent"]=="Asia"].to_crs('ESRI:54012').plot(
    ax=axs["map"],
    facecolor=bright_green,
    edgecolor=(1, 1, 1, 0)
)

axs["map"].set_axis_off()

# Letters plot
axs["letters"].set_title(
    "Letters in country's name",
    fontsize=20,
    weight="bold"
)
axs["letters"].imshow(
    df_asia.transpose(),
    aspect="auto",
    vmin=0,
    vmax=len(my_cmap.colors),
    cmap= my_cmap
)

# Major ticks
axs["letters"].set_yticks(
    ticks=range(df_asia.shape[1]),
    labels=df_asia.columns
)
axs["letters"].yaxis.tick_right()

axs["letters"].set_xticks(
    ticks=range(df_asia.shape[0]),
    labels=df_asia.index,
    rotation=90,
    ha="center"
)
# Minor ticks
axs["letters"].set_xticks(
    minor=True,
    ticks=np.arange(
        axs["letters"].get_xlim()[0]+1,
        axs["letters"].get_xlim()[1],
        1
    )
)
axs["letters"].set_yticks(
    minor=True,
    ticks=np.arange(
        axs["letters"].get_ylim()[0]-1,
        axs["letters"].get_ylim()[1],
        -1
    )
)

# Distribution plot
axs["hist"].set_title(
    "Letter distribution",
    fontsize=20,
    weight="bold"
)
axs["hist"].sharey(axs["letters"])
axs["hist"].tick_params(
    labelleft=False
)

axs["hist"].barh(
    df_asia.sum().index,
    df_asia.sum(),
    height=1,
    facecolor=bright_green,
    edgecolor=dark_green,
    lw=1
)

# Sorted distribution plot
axs["bars"].set_title(
    "Letter distribution sorted",
    fontsize=20,
    weight="bold"
)
axs["bars"].set_xticks(
    ticks=x_bar_chart_values,
    labels=[f"{val}" if i%2==0 else "" for i, val in enumerate(x_bar_chart_values)]
)
axs["bars"].set_yticks(
    ticks=range(df_asia.shape[1]),
    labels=df_asia.sum().sort_values(ascending=False).index
)
axs["bars"].set_ylim(
    axs["hist"].get_ylim()
)

axs["bars"].barh(
    df_asia.sum().sort_values(ascending=False).index,
    df_asia.sum().sort_values(ascending=False),
    height=1,
    facecolor=bright_green,
    edgecolor=dark_green,
    lw=1
)
# I need to create a 2nd axis to distribute xaxis labels better
ax_twin = axs["bars"].twiny()
ax_twin.xaxis.tick_bottom()
ax_twin.spines.bottom.set_position(("axes", -0.035))
ax_twin.set_frame_on(False)
ax_twin.set_xlim(
    axs["bars"].get_xlim()
)
ax_twin.set_xticks(
    ticks=x_bar_chart_values[1::2]
)
ax_twin.tick_params(
    bottom=False
)

# Settings
for ax in axs:
    axs[ax].set_frame_on(False)

# Ticks
axs["letters"].tick_params(
    which="both",
    bottom=False,
    right=False,
)
axs["hist"].tick_params(
    which="both",
    left=False,
)
axs["bars"].tick_params(
    which="both",
    left=False,
)

# Grids
axs["letters"].grid(
    which="minor",
    axis="both",
    color=dark_green,
    lw=1
)
axs["hist"].grid(
    which="major",
    axis="x",
    color=dark_green,
    lw=1
)
axs["hist"].set_axisbelow(True)
axs["bars"].grid(
    which="major",
    axis="x",
    color=dark_green,
    lw=1
)
axs["bars"].set_axisbelow(True)

axs["bars"].text(
    1, 0.-0.2,
    "by @pawjast\nmade with @matplotlib and @geopandas \u2665",
    transform =axs["bars"].transAxes,
    horizontalalignment="right",
    verticalalignment="bottom",
    color="#9a9996",
    # fontsize=16,
    family="monospace",
    weight="medium"
)
axs["bars"].text(
    1, 0.-0.26,
    "Source: https://www.worldometers.info/geography/how-many-countries-in-asia/",
    transform =axs["bars"].transAxes,
    horizontalalignment="right",
    verticalalignment="bottom",
    color="#9a9996",
    fontsize=10,
    family="monospace",
    weight="medium"
);

# plt.savefig(
#     fname="pawjast.png",
#     facecolor="white",
#     format="png",
#     dpi=125,
#     metadata=metadata_dict
# );
```
