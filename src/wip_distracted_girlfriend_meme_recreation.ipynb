```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import ListedColormap
```

## From Gemini

```python
# Representing the image as a 2D grid (y-axis is 30 units, x-axis is 42 units)
# This is a simplified schematic of the pixel art in the image.
data = np.zeros((30, 42))

# --- RED FIGURE (Girlfriend) ---
data[0:11, 5:19] = 1   # Dress
data[11:22, 10:15] = 4 # Face
data[11:24, 8:10] = 5  # Hair left
data[11:24, 15:17] = 5 # Hair right
data[22:24, 10:15] = 5 # Hair top

# --- BLUE FIGURE (Boyfriend) ---
data[0:20, 23:33] = 2  # Shirt
data[19:25, 23:28] = 4 # Face
data[22:25, 27:29] = 5 # Hair
data[0:10, 29:33] = 4  # Arm

# --- GREY FIGURE (Other Woman) ---
data[0:14, 35:42] = 3  # Dress/Body
data[11:19, 36:40] = 4 # Face
data[16:21, 35:41] = 5 # Hair

# Define the colors to match the meme
cmap = ListedColormap(['white', '#FF0000', '#1F618D', '#A6ACAF', '#F5B7B1', '#42312C'])

plt.figure(figsize=(10, 6))
plt.pcolor(data, cmap=cmap, edgecolors='k', linewidths=0.5)

# Formatting to look like the Excel chart
plt.title("Chart Title")
plt.xticks(range(1, 43))
plt.yticks(range(0, 31, 5))
plt.grid(True, color='grey', linestyle='-', linewidth=0.5)
plt.show()
```
